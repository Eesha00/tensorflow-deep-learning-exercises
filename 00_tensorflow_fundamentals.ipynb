```python
# Import TensorFlow
import tensorflow as tf
print(tf.__version__) # find the version number (should be 2.x+)
```

```python
# Create a scalar (rank 0 tensor)
scalar = tf.constant(1)
scalar # An empty tuple ()
         #indicates that it's a scalar (a single value with no dimensions)
```

```python
# Check the number of dimensions of a tensor (ndim stands for number of dimensions)
scalar.ndim
```

```python
# Create a vector (more than 0 dimensions)
vector = tf.constant([10, 10])
vector
#output:shape=(2,):
#It means the tensor is a 1-dimensional array (a vector) with 2 elements.
# The number inside the parentheses indicates the size of that dimension.
# If it were a 2x2 matrix, the shape would be (2, 2)
```

```python
vector = tf.constant([10,2,3])
vector
```

```python
vector.ndim
```

```python
# Create a matrix (more than 1 dimension)
matrix = tf.constant([[10, 7],
                      [7, 10]])
matrix
```

```python
# Create the same tensor with tf.Variable() and tf.constant()
changeable_tensor = tf.Variable([10, 7])
unchangeable_tensor = tf.constant([10, 7])
changeable_tensor, unchangeable_tensor
```

```python
# Won't error
changeable_tensor[0].assign(7)
changeable_tensor
```
