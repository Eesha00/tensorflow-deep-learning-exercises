```python
# Import TensorFlow
import tensorflow as tf
print(tf.__version__) # find the version number (should be 2.x+)
```

```python
# Create a scalar (rank 0 tensor)
scalar = tf.constant(1)
scalar # An empty tuple ()
         #indicates that it's a scalar (a single value with no dimensions)
```

```python
# Check the number of dimensions of a tensor (ndim stands for number of dimensions)
scalar.ndim
```

```python
# Create a vector (more than 0 dimensions)
vector = tf.constant([10, 10])
vector
#output:shape=(2,):
#It means the tensor is a 1-dimensional array (a vector) with 2 elements.
# The number inside the parentheses indicates the size of that dimension.
# If it were a 2x2 matrix, the shape would be (2, 2)
```

```python
vector = tf.constant([10,2,3])
vector
```

```python
vector.ndim
```

```python
# Create a matrix (more than 1 dimension)
matrix = tf.constant([[10, 7],
                      [7, 10]])
matrix
```

```python
# Create the same tensor with tf.Variable() and tf.constant()
changeable_tensor = tf.Variable([10, 7])
unchangeable_tensor = tf.constant([10, 7])
changeable_tensor, unchangeable_tensor
```

```python
# Won't error
changeable_tensor[0].assign(7)
changeable_tensor
```

```python
# Shuffle a tensor (valuable for when you want to shuffle your data)
not_shuffled = tf.constant([[10, 7],
                            [3, 4],
                            [2, 5]])
# Gets different results each time
tf.random.shuffle(not_shuffled)
```

```python
tf.ones(shape=(3,3))
```

```python
# Create a rank 4 tensor (4 dimensions)
rank_4_tensor = tf.zeros([2, 3, 4, 5])
rank_4_tensor
```

```python
# You can add values to a tensor using the addition operator
tensor = tf.constant([[10, 7], [3, 4]])
tensor + 10
```

```python
# Matrix multiplication in TensorFlow
print(tensor)
tf.matmul(tensor, tensor)
```

```python
X = tf.constant([[3,4,1],
                [1,2,3]])
Y = tf.constant([[1],
                 [2],
                 [3]])
X@Y
```

```python
# Perform matrix multiplication between X and Y (transposed)
tf.matmul(tf.transpose(Y), tf.transpose(X))
```

```python
import numpy as np
# Create a tensor with 50 random values between 0 and 100
E = tf.constant(np.random.randint(low=0, high=100, size=10))
E
```

```python
tf.reduce_min(E)
```

```python
# Create a rank 5 (5 dimensions) tensor of 50 numbers between 0 and 100
G = tf.constant(np.random.randint(0, 100, 50), shape=(1, 1, 1, 1, 50))
G.shape, G.ndim
```

```python
# Squeeze tensor G (remove all 1 dimensions)
G_squeezed = tf.squeeze(G)
G_squeezed.shape, G_squeezed.ndim
```

```python
some_list = [1,2,3,4]
tf.one_hot(some_list, depth = 3)
```

```python
A = tf.constant([[1, 2],
                 [3, 4]])
B = tf.constant([[5, 6],
                 [7, 8]])

# Perform matrix dot multiplication
C = tf.matmul(A, B)
print("Matrix A:\n", A.numpy())
print("Matrix B:\n", B.numpy())
print("Result of A . B:\n", C.numpy())
```

```python
vector1 = tf.constant([1, 2, 3])
vector2 = tf.constant([4, 5, 6])

# Calculate the dot product using tf.tensordot
# For two 1D vectors, tensordot with axes=1 performs the dot product
dot_product_tensordot = tf.tensordot(vector1, vector2, axes=1)

print("Vector 1:", vector1.numpy())
print("Vector 2:", vector2.numpy())
print("Dot Product (tf.tensordot):", dot_product_tensordot.numpy())
```

```python
scalar_tensor = tf.constant(42)
vector_tensor = tf.constant([1, 2, 3, 4])
matrix_tensor = tf.constant([[5, 6], [7, 8]])
general_tensor = tf.constant([[[10, 11], [12, 13]], [[14, 15], [16, 17]]])

print("Scalar Tensor:", scalar_tensor)
print("Vector Tensor:", vector_tensor)
print("Matrix Tensor:", matrix_tensor)
print("General Tensor (Rank 3):", general_tensor)
```
